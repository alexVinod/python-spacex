```python
students_df1 = pd.read_csv("students1.csv")
students_df2 = pd.read_csv("students2.csv")

std1 = students_df1.add_prefix("std1_")
std2 = students_df2.add_prefix("std2_")



merge_df = pd.merge(std1,std2, how="outer", left_on="std1_name", right_on="std2_name")
merge_df
# first_df = merge_df[merge_df[""]]
# ['std1_id', 'std1_name', 'std1_dob', 'std1_department', 'std1_mobile',
# 'std1_created_on', 'std2_id', 'std2_name', 'std2_dob',
#     'std2_department', 'std2_mobile', 'std2_created_on']
first_df = merge_df[['std1_id', 'std1_name', 'std1_dob', 'std1_department', 'std1_mobile','std1_created_on']]
second_df = merge_df[['std2_id', 'std2_name', 'std2_dob', 'std2_department', 'std2_mobile', 'std2_created_on']]

print(first_df.shape[0], second_df.shape[0])

first_df
second_df
one_two_df = first_df.merge(second_df, left_on=['std1_name', 'std1_dob', 'std1_department'], 
                           right_on=['std2_name', 'std2_dob', 'std2_department'], how='left')

def compare_data(df):
    if ((df['std1_name'] != df['std2_name'] or df['std1_name'] != "") and
        (df['std1_dob'] != df['std2_dob'] or df['std1_dob'] != "") and 
        (df['std1_department'] != df['std2_department'] or df['std1_department'] != "")):
        return 1
    else:
        return 0
one_two_df["compared"] = one_two_df.apply(compare_data, axis=1)

one_two_df
```

```python
import datetime
start = datetime.datetime.now()
print(f">>> Start: {start}")

server="localhost\MSSQLSERVER02" # localhost\MSSQLSERVER01
database="testdatabase"
try:
    stu_df = pd.read_csv("files/marvel_dc_characters1.csv")
    stu_df.dropna(inplace=True)
    stu_df["name"] = list(map(str, stu_df["name"].values.tolist() ))
    data  = stu_df.values.tolist()
    
    # define our connection string
    cnxn = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};\
                           SERVER='+ server +'; \
                           DATABASE='+ database +'; \
                           Trusted_Connection=yes;')
    
    # create the connection cursor
    cursor = cnxn.cursor()
    cursor.fast_executemany = True
    
    # definer our insert query
    #insert_query = '''INSERT INTO users(id,name,phone) VALUES (1,"Vinod","9701550994")'''
    
    rows = list(map(tuple, data))
    
    insert_query ="""
        INSERT INTO utests (id, phone, name) VALUES (?, ?, ?);
    """
    
    cursor.executemany(insert_query, rows)
    
    #sel_query = """SELECT * FROM users"""
    #result = cursor.execute(sel_query).fetchall()
    #print(result)
    
    # Commit the inserts
    cnxn.commit()    
except Exception as e:
    print(e)
    
end = datetime.datetime.now()
print(f">>> End: {end}")
finish = end - start
print(f">>> Finish: {finish} ")

# without fast_executemany == 01:19.596475
# with fast_executemany == 0:00:10.170284
```
