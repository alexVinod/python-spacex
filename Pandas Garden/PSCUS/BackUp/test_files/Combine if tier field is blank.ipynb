```python
path = r'sample_test.xlsx'
wb_obj = openpyxl.load_workbook(path)
sheet_obj = wb_obj.active
```

```python
def dataset_to_dataframe():
    """This method is used to convert the dataset dictionary to dataframe"""
    
    dataset={}
    for i, row_cells in enumerate(sheet_obj.iter_rows()):
        for cell in row_cells:
            if i in dataset.keys():
                dataset[i].append(cell.value)
            else:
                dataset[i]=[cell.value]
    return pd.DataFrame(dataset).T
```
